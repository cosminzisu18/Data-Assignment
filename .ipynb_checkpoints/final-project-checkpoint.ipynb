```python
def clean_name(name):
    # Elimină numerele și caracterele speciale, păstrând doar literele și '&'
    name = re.sub(r'[^a-zA-Zàâçéèêëîïôûùüÿñæœ&\s]', '', name)
    
    # Elimină sufixele precum 'itd', 'co', 'company', 'ltd', 'inc'
    name = re.sub(r'\b(itd|co|company|ltd|inc)\b', '', name, flags=re.IGNORECASE)
    
    # Curăță spațiile în exces
    name = name.strip()
    
    return name
```

```python
df1['name'] = df1['name'].apply(clean_name)
```

```python
def clean_address(address):
    if isinstance(address, str):  # Verifică dacă este string
        # Păstrează doar litere și cifre (fără alte caractere speciale)
        address = re.sub(r'[^a-zA-Z0-9\s]', '', address)
        # Curăță spațiile în exces
        return address.strip()
    return address  # Returnează valoarea originală dacă nu este string
```

```python
df1['address'] = df1['address'].apply(clean_address)
```

```python
df1['phone'] = df1['phone'].astype(str)
```

```python
def clean_phone(phone):
    if isinstance(phone, str):  # Verifică dacă este string
        # Extrage doar cifrele
        digits = re.sub(r'\D', '', phone)
        
        # Verifică dacă numărul are minim 11 cifre
        if len(digits) < 11:
            return ''  # Returnează spațiu gol pentru numere mai mici de 11 cifre

        # Formatează numărul
        formatted_phone = f'+{digits[:2]} {digits[2:5]}-{digits[5:8]}-{digits[8:]}'
        return formatted_phone
    return phone  # Returnează valoarea originală dacă nu este string
```

```python
df1['phone'] = df1['phone'].apply(clean_phone)
```

```python
def combine_columns(val_x, val_y):
    # Funcția va accepta două coloane ca parametri
    if pd.isna(val_x) and pd.isna(val_y):
        return ''  # Dacă ambele sunt NaN, returnăm un string gol
    
    # Transformăm valorile None sau NaN în string gol
    val_x = '' if pd.isna(val_x) else val_x
    val_y = '' if pd.isna(val_y) else val_y

    # Separăm valorile prin virgulă sau " | " și creăm o listă
    items_x = re.split(r'[,\|]', val_x)
    items_y = re.split(r'[,\|]', val_y)

    # Eliminăm spațiile albe în plus și elementele goale din listă
    items_x = [item.strip() for item in items_x if item.strip()]
    items_y = [item.strip() for item in items_y if item.strip()]

    combined_items = []
    
    # Verificăm potrivirile fuzzy și combinăm
    for item_x in items_x:
        found_match = False
        for item_y in items_y:
            # Comparăm elementele cu fuzzy matching (80% similaritate)
            if fuzz.ratio(item_x.lower(), item_y.lower()) >= 80:
                if item_x not in combined_items:
                    combined_items.append(item_x)
                found_match = True
                break
        if not found_match and item_x not in combined_items:
            combined_items.append(item_x)
    
    # Adăugăm elementele din y care nu au fost potrivite
    for item_y in items_y:
        if item_y not in combined_items:
            combined_items.append(item_y)
    
    # Returnăm rezultatul, separate prin virgulă
    return ', '.join(combined_items)
```

```python
df12['category_combined'] = df12.apply(lambda row: combine_columns(row['category_x'], row['category_y']), axis=1)
```

```python
def combine_phones(phone_x, phone_y):
    # Verificăm similaritatea între două numere de telefon folosind fuzzywuzzy
    def are_similar(phone_x, phone_y):
        return fuzz.ratio(phone_x, phone_y) >= 80

    # Dacă ambele sunt goale, returnăm șir gol
    if not phone_x and not phone_y:
        return ''
    
    # Dacă unul dintre ele este gol, returnăm celălalt
    if not phone_x:
        return phone_y
    if not phone_y:
        return phone_x

    # Dacă sunt similare (>= 80%), returnăm cel mai lung număr
    if are_similar(phone_x, phone_y):
        return phone_x if len(phone_x) >= len(phone_y) else phone_y
    
    # Dacă sunt diferite, le returnăm separate prin virgulă
    return f'{phone_x}, {phone_y}'
```

```python
df12['phone_combined'] = df12.apply(lambda row: combine_phones(row['phone_x'], row['phone_y']), axis=1)
```

```python
def combine_addresses(row):
    # Extragem adresele
    address_x = row['address_text_x']
    address_y = row['address_text_y']
    
    # Împărțim adresele în componente
    components_x = [component.strip() for component in address_x.split(',') if component.strip()]
    components_y = [component.strip() for component in address_y.split(',') if component.strip()]

    # Combinăm componentele
    combined_address = []
    
    # Asigurăm că avem aceeași lungime pentru ambele liste
    max_length = max(len(components_x), len(components_y))
    
    for i in range(max_length):
        part_x = components_x[i] if i < len(components_x) else ""
        part_y = components_y[i] if i < len(components_y) else ""
        
        if part_x == part_y and part_x:  # Păstrează dacă sunt egale și nu sunt goale
            combined_address.append(part_x)  # Adăugăm elementul comun
        elif part_x and part_y:  # Adăugăm doar dacă ambele sunt nenule
            combined_address.append(part_x + " | " + part_y)  # Adăugăm elementele diferite
        elif part_x:  # Păstrează doar componenta din x dacă y e gol
            combined_address.append(part_x)
        elif part_y:  # Păstrează doar componenta din y dacă x e gol
            combined_address.append(part_y)

    # Formăm adresa combinată
    return ', '.join(combined_address)
```

```python
df12['combined_address'] = df12.apply(combine_addresses, axis=1)
```
